```python
import os
os.environ['MKL_THREADING_LAYER'] = 'GNU'
config_file = '/mp/nas2/ruihan/computer_vision/mmediting/configs/video_interpolators/cain/cain_b5_320k_vimeo-triplet.py'
checkpoint_file = 'https://download.openmmlab.com/mmediting/video_interpolators/cain/cain_b5_320k_vimeo-triple_20220117-647f3de2.pth'
input_video = '/mp/nas2/ruihan/computer_vision/mmediting/output/qing_cain_input/' 
output_dir = '/mp/nas2/ruihan/computer_vision/mmediting/output/qing_cain/'

cmd = 'python /mp/nas2/ruihan/computer_vision/mmediting/demo/video_interpolation_demo.py  {} {} {} {} --fps_multiplier 4'.format(
    config_file,
    checkpoint_file,
    input_video,
    output_dir)
os.system(cmd)
```

```python
image_folder = '/mp/nas2/ruihan/computer_vision/mmediting/output/qing_cain/' 
video_name = '/mp/nas2/ruihan/computer_vision/mmediting/output/qing_cain/qing_cain.mp4'
images = [img for img in os.listdir(image_folder) if img.endswith(".png")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
video = cv2.VideoWriter(video_name, fourcc, 6, (width,height))
for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

cv2.destroyAllWindows()
video.release()
```
